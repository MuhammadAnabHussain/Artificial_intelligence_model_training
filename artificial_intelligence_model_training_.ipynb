```python
def matrix_product(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0])
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0])

    if cols_a != rows_b:
        raise ValueError("Number of columns in matrix_a must be equal to the number of rows in matrix_b")

    result = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]

    return result

matrix_a = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
matrix_b = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

matrix_c = matrix_product(matrix_a, matrix_b)

for row in matrix_c:
    print(row)
```

```python
import numpy as np
import pandas as pd
import random
df = pd.read_csv("/content/drive/MyDrive/datasets/diabetes (1).csv")
data = df.values

k = 6
max_iterations = 50
```

```python
def initialize_centroids(data, k):
    centroids = []
    for _ in range(k):
        centroid = random.choice(data)
        centroids.append(centroid)
    return centroids
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))
```

```python
def assign_clusters(data, centroids):
    clusters = [[] for _ in range(k)]
    for point in data:
        distances = [euclidean_distance(point, centroid) for centroid in centroids]
        closest_centroid = np.argmin(distances)
        clusters[closest_centroid].append(point)
    return clusters
```

```python
def update_centroids(clusters):
    centroids = []
    for cluster in clusters:
        cluster_mean = np.mean(cluster, axis=0)
        centroids.append(cluster_mean)
    return centroids
```

```python
def k_means(data, k, max_iterations):
    centroids = initialize_centroids(data, k)
    for _ in range(max_iterations):
        clusters = assign_clusters(data, centroids)
        new_centroids = update_centroids(clusters)


        if np.array_equal(centroids, new_centroids):
            break

        centroids = new_centroids

    return clusters, centroids
```

```python
clusters, centroids = k_means(data, k, max_iterations)


for i, cluster in enumerate(clusters):
    print(f"Cluster {i + 1}:")
    for point in cluster:
        print(point)
    print()
```
